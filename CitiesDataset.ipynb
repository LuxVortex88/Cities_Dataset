```python
import pandas as pd


data = {'Cities': ['New York', 'Los Angeles', 'Hawaii', 'Florida'],
        'Revenue': [19000, 15000, 22000, 80000]}
cities_data = pd.DataFrame(data)
display(cities_data.head())
```

```python
from matplotlib import pyplot as plt
_df_0 = cities_data.copy()
_df_0['Revenue'].plot(kind='hist', bins=20, title='Revenue')
plt.xlabel('Revenue')
plt.ylabel('Frequency')
plt.show()
```
